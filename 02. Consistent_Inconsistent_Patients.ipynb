```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster
import scipy.cluster.hierarchy as sch
from sklearn.manifold import TSNE
```

```python
#Read the preprocessed dataset
dfread = pd.read_csv("patients/patients.csv", low_memory=False).drop(columns = ['Unnamed: 0'])
df = dfread.drop(columns  = ['PPG_united', 'PPG_skew', 'PPG_kurt', 'PPG_period'])

#Make DATE in datetime format
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y-%m-%d').dt.date
```

### Feature Extraction in order to Cluster the Frequent Pilot Users

```python
#Build a dataframe with: 1. mean of day differences, 2. total days, 3. total readings, 4. total days per total readings
df2 = pd.DataFrame({'PATIENT_CODE' : df.PATIENT_CODE.unique()})
for i in range(0, len(df['PATIENT_CODE'].unique())):
    dftemp = df[df.PATIENT_CODE == df.PATIENT_CODE.unique()[i]]
    dftemp.reset_index(inplace=True)
    #Remove pilot patients with readings in only one day
    if len(df[df.PATIENT_CODE == df.PATIENT_CODE.unique()[i]]) > 1:
        k=1
        for j in range(1,len(dftemp.index)):
            if (dftemp['DATE'].iloc[j] - dftemp['DATE'].iloc[j-1]).days != 0:
                dftemp.at[j-1, 'Day differences mean'] = (dftemp['DATE'].iloc[j] - dftemp['DATE'].iloc[j-1]).days
                k += 1
        print i
        if k == 1:
            dftemp.at[j, 'Day differences mean'] = 0
        if df2.at[i, 'PATIENT_CODE'] == df.PATIENT_CODE.unique()[i]:
            df2.at[i, 'CATEGORY'] = dftemp.at[1, 'CATEGORY']
            df2.at[i, 'Day differences mean'] = dftemp['Day differences mean'].mean()
            df2.at[i, 'Total days'] = k
            df2.at[i, 'Total readings'] = len(dftemp.index)

#Produce feature 4. Total days per Total readings  
df2['Total days per Total readings'] = df2['Total days']/df2['Total readings']
#df2['Total days per Day differences mean'] = df2['Total days']/df2['Day differences mean']
#df2['Total readings per Day differences mean'] = df2['Total readings']/df2['Day differences mean']
```

```python
#Check for outliers in feature 1: Day differences mean, using z-score
df2['zscore'] = (df2['Day differences mean'] - df2['Day differences mean'].mean())/df2['Day differences mean'].std(ddof=0)
plt.plot(df2[np.abs(df2['zscore']) <= 3]['Day differences mean'], "*")
plt.plot(df2[np.abs(df2['zscore']) > 3]['Day differences mean'], "*", c='r')
plt.title('Outliers in Feature 1: Day Differences Mean')
plt.savefig('patients/outliers/Outliers_feature_1.svg', format='svg', dpi=600, bbox_inches = "tight")
plt.show()

#Remove outliers for feature 1: Day differences mean
df2.loc[(df2['zscore'] > 3) | (df2['zscore'] < -3),'zscore'] = np.nan
df2 = df2.dropna(thresh=7)
```

```python
#Check for outliers in feature 4: Total days per Total readings, using z-score
df2['zscore'] = (df2['Total days per Total readings'] - df2['Total days per Total readings'].mean())/df2['Total days per Total readings'].std(ddof=0)
plt.plot(df2[np.abs(df2['zscore']) <= 3]['Total days per Total readings'], "*")
plt.plot(df2[np.abs(df2['zscore']) > 3]['Total days per Total readings'], "*", c='r')
plt.title('Outliers in Feature 4: Ratio')
plt.savefig('patients/outliers/Outliers_feature_4.svg', format='svg', dpi=600, bbox_inches = "tight")
plt.show()

#Remove outliers for feature 4: Total days per Total readings
df2.loc[(df2['zscore'] > 3) | (df2['zscore'] < -3),'zscore'] = np.nan
df2 = df2.dropna(thresh=7)
df2 = df2.drop(columns  = [ 'zscore'])
```

### Clustering of Frequent Pilot Users with k-Means

```python
#Normalize columns
df3 = df2.drop(columns  = ['PATIENT_CODE', 'CATEGORY'])
df3=(df3-df3.mean())/df3.std()
```

```python
#Setup a kmeans classifier with 2 neighbors
kmeans = cluster.KMeans(2)

#Fit the model
kmeans.fit(df3)
cent = kmeans.cluster_centers_

#Print clustering results
print kmeans.labels_
df2['CLASSES_kmeans'] = kmeans.labels_
```

```python
#Plot the clusters obtained using k means for all combinations of the 4 features

plt.rcParams.update({'font.size': 4})
fig = plt.figure()
ax = fig.add_subplot(4,3,1)
scatter = ax.scatter(df2['Day differences mean'], df2['Total readings'], c = kmeans.labels_)
ax.set_xlabel('Day differences mean')
ax.set_ylabel('Total readings')
ax = fig.add_subplot(4,3,2)
scatter = ax.scatter(df2['Day differences mean'], df2['Total days'], c = kmeans.labels_)
ax.set_xlabel('Day differences mean')
ax.set_ylabel('Total days')
ax = fig.add_subplot(4,3,3)
scatter = ax.scatter(df2['Day differences mean'], df2['Total days per Total readings'], c = kmeans.labels_)
ax.set_xlabel('Day differences mean')
ax.set_ylabel('Total days per Total readings')
ax = fig.add_subplot(4,3,4)
scatter = ax.scatter(df2['Total readings'], df2['Total days'], c = kmeans.labels_)
ax.set_xlabel('Total readings')
ax.set_ylabel('Total days')
ax = fig.add_subplot(4,3,5)
scatter = ax.scatter(df2['Total readings'], df2['Day differences mean'], c = kmeans.labels_)
ax.set_xlabel('Total readings')
ax.set_ylabel('Day differences mean')
ax = fig.add_subplot(4,3,6)
scatter = ax.scatter(df2['Total readings'], df2['Total days per Total readings'], c = kmeans.labels_)
ax.set_xlabel('Total readings')
ax.set_ylabel('Total days per Total readings')
ax = fig.add_subplot(4,3,7)
scatter = ax.scatter(df2['Total days'], df2['Total readings'], c = kmeans.labels_)
ax.set_xlabel('Day days')
ax.set_ylabel('Total readings')
ax = fig.add_subplot(4,3,8)
scatter = ax.scatter(df2['Total days'], df2['Day differences mean'], c = kmeans.labels_)
ax.set_xlabel('Day days')
ax.set_ylabel('Day differences mean')
ax = fig.add_subplot(4,3,9)
scatter = ax.scatter(df2['Total days'], df2['Total days per Total readings'], c = kmeans.labels_)
ax.set_xlabel('Day days')
ax.set_ylabel('Total days per Total readings')
ax = fig.add_subplot(4,3,10)
scatter = ax.scatter(df2['Total days per Total readings'], df2['Total readings'], c = kmeans.labels_)
ax.set_xlabel('Total days per Total readings')
ax.set_ylabel('Total readings')
ax = fig.add_subplot(4,3,11)
scatter = ax.scatter(df2['Total days per Total readings'], df2['Day differences mean'], c = kmeans.labels_)
ax.set_xlabel('Total days per Total readings')
ax.set_ylabel('Day differences mean')
ax = fig.add_subplot(4,3,12)
scatter = ax.scatter(df2['Total days per Total readings'], df2['Total days'], c = kmeans.labels_)
ax.set_xlabel('Total days per Total readings')
ax.set_ylabel('Total days')
plt.suptitle('kMeans Clustering Results', fontsize=12, y=1.08)
plt.tight_layout()
plt.savefig('patients/clustering/kmeans_clustering.svg', format='svg', dpi=600, bbox_inches = "tight")
plt.show()
```

### Clustering Frequent Pilot Users with Hierarchical Clustering

```python
#Perform hierarchical clustering on samples using the linkage() function with the method='complete' keyword argument.
mergings = sch.linkage(df3, method='complete')

#Plot a dendrogram using the dendrogram() function on mergings, specifying the keyword arguments labels=varieties, leaf_rotation=90, and leaf_font_size=6.
sch.dendrogram(mergings, leaf_rotation=90, leaf_font_size=6)
plt.rcParams.update({'font.size': 10})
plt.title('Hierarchical clustering', fontsize=12)
plt.savefig('patients/clustering/Hierarchical_clustering.svg', format='svg', dpi=600, bbox_inches = "tight")
plt.show()

#Print the clusters obtained using Hierarchical clustering
print sch.fcluster(mergings, 2, 'maxclust')
df2['CLASSES_hierarchy'] = sch.fcluster(mergings, 2, 'maxclust')
```
